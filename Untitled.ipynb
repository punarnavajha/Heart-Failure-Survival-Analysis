```python
from ucimlrepo import fetch_ucirepo 
import pandas as pd

# Fetch dataset 
heart_failure_clinical_records = fetch_ucirepo(id=519) 

# Combine features (X) and targets (y) into one DataFrame
df = pd.concat([heart_failure_clinical_records.data.features, 
                heart_failure_clinical_records.data.targets], axis=1)

# Review the data structure
print(df.info())
df.head()
```

```python
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# Let's find the actual name of the event column automatically
# It's usually 'death_event' or 'DEATH_EVENT'
possible_names = ['death_event', 'DEATH_EVENT']
event_col = [c for c in df.columns if c.lower() == 'death_event'][0]
time_col = [c for c in df.columns if c.lower() == 'time'][0]

kmf = KaplanMeierFitter()

# Separate groups (assuming the 'smoking' column name is correct)
smokers = df[df['smoking'] == 1]
non_smokers = df[df['smoking'] == 0]

plt.figure(figsize=(10, 6))

# Fit and plot for Smokers
kmf.fit(smokers[time_col], event_observed=smokers[event_col], label='Smokers')
kmf.plot_survival_function()

# Fit and plot for Non-Smokers
kmf.fit(non_smokers[time_col], event_observed=non_smokers[event_col], label='Non-Smokers')
kmf.plot_survival_function()

plt.title(f'Survival Probability: Smokers vs Non-Smokers')
plt.xlabel('Days of Follow-up')
plt.ylabel('Survival Probability')
plt.grid(True)
plt.show()
```

```python
from lifelines import CoxPHFitter

# This logic automatically finds the columns even if they are lowercase
event_col = [c for c in df.columns if c.lower() == 'death_event'][0]
time_col = [c for c in df.columns if c.lower() == 'time'][0]

# Select your predictors. We use .lower() check to be safe for all of them.
# Typical UCI names: 'age', 'ejection_fraction', 'serum_creatinine'
cols_to_use = [
    'age', 'ejection_fraction', 'serum_creatinine', 
    time_col, event_col
]

# Create a clean dataframe for the model
cph_df = df[cols_to_use]

# Run the model
cph = CoxPHFitter()
cph.fit(cph_df, duration_col=time_col, event_col=event_col)

# Display the results
cph.print_summary()
```

```python
# Check the Proportional Hazards assumption
cph.check_assumptions(cph_df, p_value_threshold=0.05)
```
